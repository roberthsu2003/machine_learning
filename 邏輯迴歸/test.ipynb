```python
import matplotlib as mpl

from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

X_cls, y_cls = make_blobs(n_samples=100, centers=2, n_features=2, random_state=1, cluster_std=1.5)
print(X_cls.shape)
print(y_cls.shape)

#繪製散點圖
plt.figure(figsize=(6, 4))
plt.scatter(X_cls[y_cls == 0][:,0],X_cls[y_cls == 0][:, 1], marker="o", color="blue", label="類別 0")
plt.scatter(X_cls[y_cls == 1][:, 0], X_cls[y_cls == 1][:, 1], marker="^", color="red", label="類別 1")

#假設一條簡單的決策邊界(僅為示意)
x_boundary = np.linspace(X_cls[:, 0].min(), X_cls[:, 0].max(), 100)
y_boundary = -0.8 * x_boundary + (-6)
plt.plot(x_boundary, y_boundary, color='green', linestyle='--', label="示意決策邊界")
plt.title("邏輯分類概念示意圖")
plt.xlabel("特徵 1")
plt.ylabel('特徵 2')
plt.legend()
plt.grid(True)
plt.show()
```
