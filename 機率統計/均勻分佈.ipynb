# 均勻分佈 (Uniform Distribution) 說明

## 1. 均勻分佈的目的

均勻分佈是一種機率分佈，其特點為：

- 在給定區間內，每個數值出現的機率都相等
- 常用於模擬隨機事件，如：
  - 隨機數生成
  - 電腦模擬
  - 排隊理論分析
  - 品質管制

## 2. 以圖表檢查均勻分佈

以下是使用 Python 繪製均勻分佈的示例：



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager
import seaborn as sns

#中文字型
fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

# 生成均勻分佈的數據
uniform_data = np.random.uniform(0, 10, 1000)

# 創建圖表
plt.figure(figsize=(10, 6))

# 繪製直方圖
sns.histplot(uniform_data, bins=30, kde=True)
plt.title('均勻分佈示例')
plt.xlabel('數值')
plt.ylabel('頻率')

# 顯示圖表
plt.show()

# 基本統計資訊
print(f"平均值: {np.mean(uniform_data):.2f}")
print(f"標準差: {np.std(uniform_data):.2f}")
print(f"最小值: {np.min(uniform_data):.2f}")
print(f"最大值: {np.max(uniform_data):.2f}")
```



這段代碼會：
1. 生成 1000 個在 [0,10] 區間內的均勻分佈隨機數
2. 繪製直方圖和核密度估計曲線
3. 輸出基本統計資訊

觀察要點：
- 直方圖應該呈現近似矩形分佈
- 各個區間的頻率應該大致相等
- 核密度估計曲線應該較為平坦
