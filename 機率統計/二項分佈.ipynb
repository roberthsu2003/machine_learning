# 二項分佈 (Binomial Distribution)

## 定義與目的
二項分佈是一個離散機率分佈，用於描述n次獨立的是/否試驗（伯努利試驗）中成功的次數。主要用途：

1. 模擬固定次數重複試驗的成功次數
2. 計算特定事件發生特定次數的機率
3. 適用於只有兩種可能結果的試驗

## 特性
- 每次試驗都是獨立的
- 每次試驗的成功機率p保持不變
- 結果只有兩種可能（成功/失敗）

## Python實作範例
以下是使用Python繪製二項分佈的示例：



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import matplotlib as mpl
from matplotlib.font_manager import fontManager

#中文字型
fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

# 設定參數
n = 20  # 試驗次數
p = 0.3 # 成功機率

# 計算機率質量函數
k = np.arange(0, n+1)
pmf = binom.pmf(k, n, p)

# 繪製圖表
plt.figure(figsize=(10, 6))
plt.bar(k, pmf)
plt.title('二項分佈 (n=20, p=0.3)')
plt.xlabel('成功次數 k')
plt.ylabel('機率')
plt.grid(True, alpha=0.3)
plt.show()
```



這個程式碼會產生一個直方圖，顯示在20次試驗中，每種可能的成功次數出現的機率。機率分佈呈現出典型的鐘形曲線，其中最可能出現的成功次數接近 n*p。

## 應用場景
- 品質管制：檢測產品缺陷
- 醫學研究：臨床試驗成功率
- 市場調查：顧客購買行為
- 生物實驗：基因遺傳特徵

讓我說明如何調整二項分佈的參數來觀察不同的分佈情況。

### 主要可調整的參數
1. `n`: 試驗次數
2. `p`: 成功機率

讓我們創建一個可以比較不同參數組合的程式碼：



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 創建2x2的子圖
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
plt.suptitle('不同參數的二項分佈比較', fontsize=16)

# 不同參數組合
params = [
    {'n': 20, 'p': 0.3, 'title': 'n=20, p=0.3'},
    {'n': 20, 'p': 0.7, 'title': 'n=20, p=0.7'},
    {'n': 50, 'p': 0.3, 'title': 'n=50, p=0.3'},
    {'n': 10, 'p': 0.5, 'title': 'n=10, p=0.5'}
]

# 繪製每個子圖
for ax, param in zip(axes.flat, params):
    n = param['n']
    p = param['p']
    k = np.arange(0, n+1)
    pmf = binom.pmf(k, n, p)

    ax.bar(k, pmf, alpha=0.8)
    ax.set_title(param['title'])
    ax.set_xlabel('成功次數 k')
    ax.set_ylabel('機率')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```



### 參數調整效果說明

1. **調整 n (試驗次數)**
   - 較大的 n：分佈更接近常態分佈
   - 較小的 n：分佈較為離散

2. **調整 p (成功機率)**
   - p = 0.5：分佈對稱
   - p < 0.5：分佈偏左
   - p > 0.5：分佈偏右

### 建議嘗試的參數組合

1. 對稱分佈：
   - n = 20, p = 0.5

2. 偏態分佈：
   - n = 20, p = 0.1（強左偏）
   - n = 20, p = 0.9（強右偏）

3. 不同試驗次數：
   - n = 100, p = 0.5（更接近常態分佈）
   - n = 10, p = 0.5（更離散的分佈）

您可以透過調整這些參數來觀察二項分佈的不同形態。
