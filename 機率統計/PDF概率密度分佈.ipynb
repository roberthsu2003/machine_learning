````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager
# 設定中文字型
fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

# 產生標準常態分布的隨機樣本
rng = np.random.default_rng()
xdata = rng.normal(size=1000)

# 建立 x 軸範圍
xpdf = np.arange(-4, 4, 0.1)

# 說明3-手動計算理論 PDF
#pdf = 1 / (np.sqrt(2 * np.pi)) * np.exp(-xpdf**2 / 2)

pdf = norm.pdf(xpdf)  # 預設為標準常態分布（平均數0，標準差1）

# 畫直方圖（density=True 代表 y 軸為機率密度）
plt.hist(xdata, bins=30, density=True, alpha=0.5, label='直方圖（密度）')

# 畫理論 PDF 曲線
plt.plot(xpdf, pdf, 'r-', label='理論 PDF')

plt.xlabel('x')
plt.ylabel('機率密度')
plt.title('常態分布 PDF')
plt.legend()
plt.show()
````

```pyhton
rng = np.random.default_rng()
xdata = rng.normal(size=1000)
```

1. `rng = np.random.default_rng()`
   - 這行程式碼是建立一個 NumPy 的亂數產生器（random number generator, RNG）實例。
   - `default_rng()` 是 NumPy 推薦的新一代亂數產生器，能提供更好的隨機性與可重現性。
   - 這樣做的好處是可以更靈活地控制亂數來源，並避免使用舊的全域亂數狀態。

2. `xdata = rng.normal(size=1000)`
   - 這行程式碼利用剛剛建立的亂數產生器 `rng`，產生 1000 筆標準常態分布（平均值為 0，標準差為 1）的隨機數據。
   - `normal()` 是產生常態分布亂數的方法，`size=1000` 表示要產生 1000 個數值。
   - 最後，這些亂數會被存放在 `xdata` 這個 NumPy 陣列中。

總結：這兩行程式碼的功能是建立一個亂數產生器，並用它產生 1000 筆標準常態分布的隨機數據。

### ⬆︎說明2
在常態分布（Normal Distribution）中，大部分的數據（約99.99%）都會落在平均值的左右4個標準差範圍內。  
因此，當我們要繪製常態分布的機率密度函數（PDF）時，通常會選擇從 -4 到 4 這個範圍，這樣可以完整呈現常態分布的主要形狀與特徵。

簡單來說，選擇 -4 到 4 之間，是因為這個範圍幾乎涵蓋了所有常態分布下可能出現的數值，能讓圖形更完整且具有代表性。

這行程式碼的說明如下：



````python
pdf = 1 / (np.sqrt(2 * np.pi)) * np.exp(-xpdf**2 / 2)
````



### 公式說明
這是**標準常態分布（Standard Normal Distribution）**的機率密度函數（Probability Density Function, PDF）公式：

\[
f(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)
\]

- `np.sqrt(2 * np.pi)`：計算分母 \(\sqrt{2\pi}\)
- `np.exp(-xpdf**2 / 2)`：計算指數部分 \(\exp(-x^2/2)\)
