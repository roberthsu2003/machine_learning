# 解釋PDF
- 以下是一個用 `matplotlib` 畫出概率密度函數（PDF）的簡單範例，這裡以標準常態分布（Normal Distribution）為例，並同時示範如何用隨機樣本畫出直方圖與理論 PDF 曲線：

```python
#下載字型
import wget
wget.download("https://github.com/roberthsu2003/machine_learning/raw/refs/heads/main/source_data/ChineseFont.ttf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager
# 設定中文字型
fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

# 產生標準常態分布的隨機樣本
rng = np.random.default_rng()
xdata = rng.normal(size=1000)

# 建立 x 軸範圍
xpdf = np.arange(-4, 4, 0.1)

# 計算理論 PDF
pdf = 1 / (np.sqrt(2 * np.pi)) * np.exp(-xpdf**2 / 2)

# 畫直方圖（density=True 代表 y 軸為機率密度）
plt.hist(xdata, bins=30, density=True, alpha=0.5, label='直方圖（密度）')

# 畫理論 PDF 曲線
plt.plot(xpdf, pdf, 'r-', label='理論 PDF')

plt.xlabel('x')
plt.ylabel('機率密度')
plt.title('常態分布 PDF')
plt.legend()
plt.show()
```
