### 欠擬合(underfitting) 與 過度擬合(overfitting)

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成模擬數據
np.random.seed(0)
x = np.linspace(0, 1, 50)  # 生成 50 個點，範圍在 0 到 1
y = x**2 + np.random.normal(0, 0.05, 50)  # 真實數據為二次函數加上噪聲

# 擬合模型
# 欠擬合：用一次多項式（直線）
underfit_coeffs = np.polyfit(x, y, 1)
underfit_model = np.poly1d(underfit_coeffs)

# 適當擬合：用二次多項式（拋物線）
goodfit_coeffs = np.polyfit(x, y, 2)
goodfit_model = np.poly1d(goodfit_coeffs)

# 過度擬合：用高次多項式（10次）
overfit_coeffs = np.polyfit(x, y, 10)
overfit_model = np.poly1d(overfit_coeffs)

# 繪製三個子圖
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 12), sharex=True, sharey=True)

# 設置圖表樣式
for ax in [ax1, ax2, ax3]:
    ax.scatter(x, y, color='blue', alpha=0.5, label='Data')  # 數據點
    ax.set_ylim(-0.2, 1.2)
    ax.grid(True, linestyle='--', alpha=0.7)

# 欠擬合圖表
ax1.plot(x, underfit_model(x), color='blue', linewidth=2, label='Underfit')
ax1.text(0.05, 1.0, 'Underfit', color='red', fontsize=12, fontweight='bold')
ax1.legend()

# 適當擬合圖表
ax2.plot(x, goodfit_model(x), color='blue', linewidth=2, label='Fit')
ax2.text(0.05, 1.0, 'Fit', color='green', fontsize=12, fontweight='bold')
ax2.legend()

# 過度擬合圖表
ax3.plot(x, overfit_model(x), color='blue', linewidth=2, label='Overfit')
ax3.text(0.05, 1.0, 'Overfit', color='orange', fontsize=12, fontweight='bold')
ax3.legend()

# 設置標籤
ax3.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_ylabel('Y')
ax3.set_ylabel('Y')

# 調整佈局
plt.tight_layout()
plt.show()
```
