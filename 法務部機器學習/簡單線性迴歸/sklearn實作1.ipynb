- 這個範例將演示如何使用 scikit-learn 的 LinearRegression 模型來預測簡單的線性關係。
- 假設我們有一組房屋面積和價格的數據，我們希望建立一個模型，根據房屋面積來預測價格。

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib as mlp
from matplotlib.font_manager import fontManager

fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')

# 1. 準備數據
# 房屋面積 (平方公尺)
X = np.array([50, 60, 70, 80, 90, 100]).reshape(-1, 1)
# 房屋價格 (百萬)
y = np.array([3, 3.5, 4, 4.5, 5, 5.5])

# 2. 建立線性迴歸模型
model = LinearRegression()

# 3. 訓練模型
model.fit(X, y)

# 4. 進行預測
# 預測 120 平方公尺的房屋價格
new_X = np.array([[120]])
predicted_y = model.predict(new_X)
print(f"預測 120 平方公尺的房屋價格為：{predicted_y[0]:.2f} 百萬")

# 5. 可視化結果
plt.scatter(X, y, color='blue', label='實際數據')
plt.plot(X, model.predict(X), color='red', label='線性迴歸模型')
plt.scatter(new_X, predicted_y, color='green', label='預測數據')
plt.xlabel('房屋面積 (平方公尺)')
plt.ylabel('房屋價格 (百萬)')
plt.title('房屋面積 vs. 價格 (線性迴歸)')
plt.legend()
plt.show()

# 6. 取得模型參數
print(f"迴歸係數 (斜率)：{model.coef_[0]:.2f}")
print(f"截距：{model.intercept_:.2f}")
```
