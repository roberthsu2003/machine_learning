```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# 載入資料
url = "Salary_Data2.csv"
data = pd.read_csv(url)
data
```

```python
# Label Encoding for EducationLevel
le = LabelEncoder()
data['EducationLevel'] = le.fit_transform(data['EducationLevel'])
data
```

```python
# OneHot Encoding for City
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False) # sparse=False for easier handling
city_encoded = ohe.fit_transform(data[['City']])
city_df = pd.DataFrame(city_encoded, columns=ohe.get_feature_names_out(['City']))
data = pd.concat([data, city_df], axis=1)
data = data.drop('City', axis=1) # Drop original City column
data
```

```python
# Prepare data for model
x = data.drop('Salary', axis=1)
y = data['Salary']
x, y
```

```python
# 分割資料
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=76) #random_state for reproducibility
x_train, x_test, y_train, y_test
```

```python
# Feature Scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train, x_test
```

```python
# 使用線型模型訓練
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# 使用測試集測試
y_pred = model.predict(x_test)
y_pred
```

```python
# 評估模型(example: R-squared)
# r2愈接近1愈好
r2 = model.score(x_test, y_test)
print(f"R-squared: {r2}")
```

```python
# 取得w(權重),b(偏移值)
w = model.coef_
b = model.intercept_
print(f"w: {model.coef_}")
print(f"b: {model.intercept_}")
```

```python



#使用模型
#預測資料: 5.3 碩士以上 城市A

x_real = np.array([[5.3, 2, 1, 0, 0]])
x_real = scaler.transform(x_real)
y_pred = (w * x_real).sum(axis=1) + b
y_pred
```
