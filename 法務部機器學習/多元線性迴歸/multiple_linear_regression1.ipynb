```python
import pandas as pd

url = "Salary_Data2.csv"
data = pd.read_csv(url)
data
```

### Label Encoding
### 將文字轉為數值
### 有高低的欄位(EducationLevel)
- 高中以下 -> 0
- 大學 -> 1
- 碩士以上 -> 2

```python
educationLabel_encoding = data['EducationLevel'].map({'高中以下':0,'大學':1,'碩士以上':2})
display(educationLabel_encoding)
```

```python
data['EducationLevel'] = educationLabel_encoding
data
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(categories=[['城市A', '城市B', '城市C']], handle_unknown='ignore') #如果出現新的城市，handle_unknown='ignore' 至關重要，以避免錯誤
#轉換過程必需是2維的資料
#由serirec轉換為DataFrame
city_encoded = onehot_encoder.fit_transform(data[['City']],)
# city_encoded是一個稀疏矩陣,轉換為nd_array
city_encoded.toarray()
```

```python
#資料加入至data內

data[['CityA','CityB','CityC']] = city_encoded.toarray()
data
```

```python
#刪除欄位
data = data.drop(['City','CityC'],axis=1)
data
```

```python
#取得所有特徵
x = data[['YearsExperience','EducationLevel','CityA','CityB']]
y = data['Salary']
```

```python
#分割數據集成為訓練集和測試集
from sklearn.model_selection import train_test_split

#random_state的目的是不要亂數分割,以利範例後面的講解,正常不需要加入
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=76) #測試集佔20%,會return 個元素的tuple,

len(x), len(x_train), len(x_test)
```

```python
#將訓練資料和測試資料轉換為nd_array

x_train = x_train.to_numpy()
x_test = x_test.to_numpy()

x_train
```

```python
#特徵縮放
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train, x_test
```

```python
#建立計算cost_function的function
def compute_cost(x, y, w, b):
    y_pred = (x_train*w).sum(axis=1) + b
    cost = ((y_train - y_pred)**2).mean()
    return cost
```

```python
#測試cost_function是否正確
import numpy as np

w = np.array([1, 2, 3, 4])
b = 1
compute_cost(x_train, y_train, w=w, b=b)
```

### 設定optimizer(gradient descent)

```python
#計算學習率

y_pred = (x_train*w).sum(axis=1) + b
b_gradient = (y_pred - y_train).mean()
#w1_gradient = x1 * (y_pred - y_train)
w1_gradient = (x_train[:,0] * (y_pred - y_train)).mean()
w2_gradient = (x_train[:,1] * (y_pred - y_train)).mean()
w3_gradient = (x_train[:,2] * (y_pred - y_train)).mean()
w4_gradient = (x_train[:,3] * (y_pred - y_train)).mean()

w1_gradient, w2_gradient, w3_gradient, w4_gradient, b_gradient
```

```python
#收集斜率的方式至w_gradient
w_gradient = np.zeros(x_test.shape[1])#x_test.shape[1]得到4
for i in range(x_train.shape[1]):
    w_gradient[i] = (x_train[:,i] * (y_pred - y_train)).mean()
```

```python
#建立function,計算斜率
def compute_gradient(x, y, w, b):
    y_pred = (x*w).sum(axis=1) + b
    w_gradient = np.zeros(x.shape[1])
    b_gradient = (y_pred - y).mean()
    for i in range(x.shape[1]):
        w_gradient[i] = (x[:,i] * (y_pred - y)).mean()
    return w_gradient, b_gradient
```

```python
#呼叫compute_gradient,計算斜率
import numpy as np

w = np.array([1, 2, 3, 4])
b = 1

compute_gradient(x_train, y_train, w, b)
```

```python
#透過學習率,改變參數
w = np.array([1, 2, 3, 4])
b = 1
learning_rate = 0.001
w_gradient, b_gradient = compute_gradient(x_train, y_train, w, b)
print("未更新前的cost")
print(compute_cost(x_train, y_train,w,b))
w = w - w_gradient * learning_rate
b = b - b_gradient * learning_rate
print("更新後的cost")
print(compute_cost(x_train, y_train,w,b))
```

```python
#將梯度下降方式寫成function
def gradient_descent(x, y, w_init, b_init, learning_rate, cost_function, gradient_function, run_iter, p_iter=1000):
  c_hist = [] #收集所有的cost
  w_hist = [] #收集所有的w
  b_hist = [] #收集所有的b

  w = w_init
  b = b_init

  for i in range(run_iter):
    w_gradient, b_gradient = gradient_function(x, y, w, b)
    w = w - w_gradient*learning_rate
    b = b - b_gradient*learning_rate
    cost = cost_function(x, y, w, b)

    w_hist.append(w)
    b_hist.append(b)
    c_hist.append(cost)

    if i%p_iter == 0:
      print(f"Iteration {i:5} : Cost {cost: .4e}, w: {w}, b: {b: .2e}, w_gradient: {w_gradient}, b_gradient: {b_gradient:.2e}")

  return w, b, w_hist, b_hist, c_hist
```

```python
w_init = np.array([1, 2, 2, 4])
b_init = 0
learning_rate = 1.0e-3
run_iter = 20000
w_final, b_final, w_hist, b_hist, c_hist = gradient_descent(x_train, y_train, w_init, b_init, learning_rate, compute_cost, compute_gradient, run_iter)
#列印出最終的w和b值
print(f"最終的w={w_final},b={b_final:.2f}")
```

```python
#測試x_test
y_pred = (w_final * x_test).sum(axis=1) + b_final
pd.DataFrame({
    "y_pred": y_pred,
    "y_test": y_test
})
```

```python
#計算測試集的cost value,然後比較訓練出來的cost value
compute_cost(x_test, y_test, w_final, b_final)
```
