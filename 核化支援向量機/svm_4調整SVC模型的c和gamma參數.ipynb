### 調整SVM參數
- 已下圖表資料已經預先設計好



```python
import matplotlib.pyplot as plt
import mglearn

fig, axes = plt.subplots(3, 3, figsize=(15, 10))

for ax, C in zip(axes, [-1, 0, 3]):
    ax
    for a, gamma in zip(ax, range(-1,2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)

axes[0, 0].legend(["class 0", "class 1", "sv class 0", "sv class 1"],ncol=4, loc=(.9, 1.2))
```

⬆︎參數 C 和 gamma 的不同設定的決策邊界和支援向量

⬆︎從左到右，我們將參數 gamma 的值從 0.1 增加到 10。較小的 gamma 意味著高斯核的半徑較大，這意味著許多點被視為接近。這反映在非常平滑的決策邊界上,左側是邊界，而右側的邊界則更注重單一點。一個較低的 gamma 值意味著決策邊界變化緩慢，從而產生低複雜度的模型，而較高的 gamma 值則產生更複雜的模型。

從上到下，我們將 C 參數從 0.1 增加到 1000。與線性模型一樣，較小的 C 意味著模型非常受限，其中每個資料點只能產生非常有限的影響。您可以看到，左上角的決策邊界看起來幾乎是線性的，錯誤分類的點幾乎對線沒有任何影響。如右下角所示，增加 C 可以讓這些點對模型產生更大的影響，並使決策邊界彎曲以正確地對它們進行分類。
